```python
unclean.drop('reference',axis=1, inplace=True)
```

```python
unclean['price'] = [str(i).replace(",", "").strip() for i in unclean['price']]
unclean['price'] = [float(str(i).replace("£", "").strip()) for i in unclean['price']]
```

```python
unclean.head()
```

```python
unclean.rename(columns={'engine size': 'tax'}, inplace=True)
unclean['tax'] = [str(i).replace("£", "").strip() for i in unclean['tax']]
unclean['tax'] = [str(i).lower().replace("cc", "").strip() for i in unclean['tax']]
unclean['tax'] = [str(i).replace(",", "").strip() for i in unclean['tax']]
unclean['tax'] = [float(str(i).strip()) for i in unclean['tax']]
unclean['tax']
```

```python
unclean.drop('fuel type', axis=1, inplace = True)
```

```python
unclean['mpg']= np.nan
```

```python
unclean.rename(columns={'fuel type2' : 'fuelType'}, inplace= True)
unclean['fuelType'].unique()
```

```python
unclean.rename(columns={'engine size2' : 'engineSize'}, inplace= True)
```

```python
unclean.head()
```

```python
unclean['engineSize'].unique()
```

```python
unclean['engineSize'] = [str(i).replace('cc','').strip()  for i in unclean['engineSize']]
unclean['engineSize'] = [str(i).replace('T','').strip()  for i in unclean['engineSize']]
unclean['engineSize'] = [str(i).replace('Unknown','').strip()   for i in unclean['engineSize']]
unclean['engineSize'] = pd.to_numeric(unclean['engineSize'], errors='coerce')
unclean['engineSize'] = unclean.apply(lambda x: x['engineSize']/1000 if x['engineSize']> 20 else x['engineSize'] , axis=1)
```

```python
unclean['engineSize'].unique()
```

```python
mileage=['mileage','mileage2']
for i in mileage:
    unclean[i] = [str(i).split('.')[0] for i in unclean[i]]
    unclean[i] = [str(i).replace(',', '') for i in unclean[i]]
    unclean[i] = [str(i).replace('Unknown', '') for i in unclean[i]]
    unclean[i]=unclean[i].replace('nan', '')
    unclean[i] = pd.to_numeric(unclean[i])
```

```python
unclean["mileage"] = unclean.apply(lambda x: x["mileage"] if x["mileage"]!= np.NaN and x["mileage"]>x["mileage2"] else x["mileage2"], axis=1)
```

```python
unclean.drop(["mileage2"], inplace=True, axis=1)
```

```python
unclean['mileage'].fillna(unclean['mileage'].mean(), inplace = True)
unclean['tax'].fillna(unclean['tax'].mean(), inplace = True)
unclean['mpg'].fillna(unclean['mpg'].mean(), inplace = True)
unclean['model'].fillna(unclean['model'].mode()[0], inplace = True)
unclean['transmission'].fillna(unclean['transmission'].mode()[0], inplace = True)
unclean['fuelType'].fillna(unclean['fuelType'].mode()[0], inplace = True)
unclean['year'].fillna(unclean['year'].mode()[0], inplace = True)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer()

unclean['price'] = imputer.fit_transform(unclean[['price']])
unclean['engineSize'] = imputer.fit_transform(unclean[['engineSize']])
```

```python
import datetime

current_year = datetime.datetime.now().year
unclean['car_age'] = current_year - unclean['year']
```

```python
unclean['fuel_engine'] = unclean['fuelType'] + '_' + unclean['engineSize'].astype(str)
```

```python
unclean['mileage'].fillna(unclean.groupby(['model', 'year'])['mileage'].transform('mean'), inplace=True)
```

```python
bins = [0, 20000, 30000, 40000, float('inf')]
labels = ['0-20k', '20k-30k', '30k-40k', '40k+']
unclean['price_range'] = pd.cut(unclean['price'], bins=bins, labels=labels, include_lowest=True)
```

```python
df = pd.get_dummies(unclean, columns=['transmission', 'fuelType'], drop_first=True)
```

```python
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Scatter plot for Price vs Mileage with color-coded fuel type
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
scatter_plot = plt.scatter(df['mileage'], df['price'], c=df['fuelType_Petrol'])
plt.title('Car Price vs Mileage by Fuel Type')
plt.xlabel('Mileage (in miles)')
plt.ylabel('Price')
plt.colorbar(scatter_plot, label='Fuel Type (Petrol)')

# Bar chart for average price by fuel type
plt.subplot(1, 2, 2)
average_price_by_fuel = df.groupby('fuelType_Petrol')['price'].mean()
bar_chart = average_price_by_fuel.plot(kind='bar', color='skyblue')
plt.title('Average Price for Petrol Fuel Type')
plt.xlabel('Fuel Type (Petrol)')
plt.ylabel('Average Price')

# Adjust layout for better organization
plt.tight_layout()

# Show the dashboard
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Set style for seaborn plots
sns.set(style="whitegrid")

# Scatter plot: Price vs Mileage colored by Fuel Type
plt.figure(figsize=(10, 6))
sns.scatterplot(x='mileage', y='price', hue='fuelType', data=unclean, palette='Set2')
plt.title('Car Price vs Mileage Colored by Fuel Type')
plt.xlabel('Mileage (in miles)')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='mileage', y='price', hue='transmission', data=unclean, palette='Set2')
plt.title('Car Price vs Mileage Colored by Fuel Type')
plt.xlabel('Mileage (in miles)')
plt.ylabel('Price')
plt.show()

# Boxplot: Price distribution by Transmission type
plt.figure(figsize=(10, 6))
sns.boxplot(x='transmission', y='price', data=unclean, palette='viridis')
plt.title('Price Distribution by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Price')
plt.show()

# Pairplot: Explore relationships between numerical variables
sns.pairplot(df[['price', 'mileage', 'engineSize', 'car_age']])
plt.suptitle('Pairplot of Numerical Variables', y=1.02)
plt.show()

# Heatmap: Correlation matrix
correlation_matrix = df[['price', 'mileage', 'engineSize', 'car_age']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
